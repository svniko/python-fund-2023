````python
import requests

city = input('Enter a city -> ')

API_key = 'c124fdaaa610d8f8091ce555740c7cd8'
p = {'appid': API_key, 'q':city}
url = f'https://api.openweathermap.org/data/2.5/weather'

response = requests.get(url, params= p).json()

if response['cod'] == 200:
    temp = round(response['main']['temp'] - 273.15, 1)
    print(f'Temperature: {temp}')
    state = response['weather'][0]['description']
    print(f'Weather: {state}')
    hum = response['main']['humidity']
    print(f'Humidity: {hum}%')
else: 
    print('Something went wrong')
````

### Using .bat file to run Python file

1) Create a .bat file (i.e. `weather.bat`)
```windows
@echo off
title Weather App

REM Це коментар у пакетному файлі
REM Наступний рядок встановлює розміри вікна

mode con: cols=40 lines=15
color 8E

py D:\py\2023_autumn\2\weather.py %1
@pause
```

Type `help color` in command prompt to find the color codes that can be used.

2) Add the folder where the `weather.bat` file is saved to the environment variables of the current user.

3) Now you can call the `weather.py` file simply from the program launch window (Win+R) by typing `weather`
